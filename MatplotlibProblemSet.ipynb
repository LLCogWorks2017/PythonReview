#### Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed. 

```python
def remove_char(old_string, n):
    front = old_string[:n]   # up to but not including n
    back = old_string[n+1:]  # n+1 through end of string
    return front + back
```

#### Test your function with three different parameters

```python
print(remove_char("radar", 2))
print(remove_char("r", 0))
print(remove_char("parameters", -2))
```

#### Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
    if len(old_string) <= 1:
        return old_string

    mid = old_string[1:-1]  

    # last + mid + first
    return old_string[len(old_string)-1] + mid + old_string[0]
```

#### Test the function with three different parameters

```python
print(swap_chars('code'))
print(swap_chars('a'))
print(swap_chars('ab'))
```
